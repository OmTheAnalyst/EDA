Through the EDA, we discovered several key findings regarding world population trends:

#### Population Growth:

Analysis of 'netChange' and 'growthRate' columns reveals the population increase or decrease and the corresponding growth rate for each country.
#Visualize population growth using line plots or bar plots for countries with the highest net change.

```python
#line plot of population over time
plt.plot(df['pop1980'],df['pop2050'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Over Time')
plt.show()
```

```python
#Top 10 countries with highest population

topten = df.nlargest(10, 'netChange')
plt.bar(topten['country'], topten['netChange'])
plt.xlabel('Country')
plt.ylabel('Net Population Change')
plt.title('Top 10 Countries with Highest Population Increases')
plt.xticks(rotation=45)
plt.show()
```

#### Regional Variations:

Explore the 'place' column to group countries by region or continent.
Analyze population distribution by region using pie charts or stacked bar plots to highlight regional variations.


```python
countries = ['India','China','United States', 'United Kingdom','Japan','France','Spain','Australia']

years = ['pop1980','pop2000','pop2010','pop2023','pop2030','pop2050']

plt.figure(figsize=(10,6))

for country in countries:
    country_data = df[df['country'] == country]
    plt.plot(years, country_data[years].values.reshape(-1), label
            =country)
    
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Trend for Selected Countries')
plt.legend()
plt.show()
```

#### Fertility Rates:

Investigate the 'place' column to analyze fertility rates by country or region.
Use box plots or line plots to visualize the distribution of 'pop2023' or 'pop2050' projections.


```python
plt.boxplot(df['density'])
plt.ylabel('Population Density')
plt.title('Box Plot of Population Density')
plt.show()
```

```python
country_name = 'India'
plt.bar(country_data.columns[14:16], country_data.iloc[:,14:16].values.flatten())
plt.xlabel('Year')
plt.ylabel('Fertility Rate')
plt.title(f'Fertility Rate Comparison for {country_name} From 2010 and 2020')
plt.show()
```
